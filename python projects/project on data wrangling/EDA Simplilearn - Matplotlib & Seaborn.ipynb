```python
import numpy as np
import pandas as pd
# matplotlib = Basic Plots made easy
import matplotlib.pyplot as plt
# Descriptive Plots made easy
import seaborn as sns
%matplotlib inline
```

```python
auto_price = pd.read_csv('E:/Data/Automobile_price_data__Raw_.csv', sep = ',')
display(auto_price)
```

```python
auto_price.dtypes
```

```python
# convert object columns to numerical columns
cols = ['bore','stroke','horsepower','peak-rpm','price']

auto_price[cols] = auto_price[cols].apply(pd.to_numeric, args = ('coerce',))
```

```python
auto_price.isnull().sum()
```

```python
# Remove/Drop the Missing Value Rows
auto_price.dropna(inplace = True)
```

#### Line Plot
        * Relationship Plot (Bivariate)
        * Trends/Values - Univariate 

```python
# Using Pandas for Visualization
auto_price['price'].plot(figsize = (15,8))
```

```python
sns.set_style('whitegrid')
plt.figure(figsize = (15,8))
plt.plot(auto_price['price'], color = 'deeppink', alpha = 0.5)
plt.title('Price of Vehicle Line Plot', size = 12)
plt.xlabel('Row Indexes')
plt.ylabel('Price')
plt.show()
```

#### Bar Plot

       * Bivariate Analysis
       * Bar chart used by Categorical, Nominal , Categorical vs. Numerical/Continious
       * Bar Plots are used to display counts of unique value of categorical dtype, Height of Bar represents count for each category.

```python
sns.set_style('whitegrid')
auto_price[['price']][10:70].plot.bar(figsize = (15,7))
```

```python
counts = auto_price['make'].value_counts()
counts
```

```python
plt.figure(figsize = (7,6))
auto_price['make'].value_counts().plot.bar(figsize = (15,7))
plt.show()
```

```python
auto_price[['city-mpg','horsepower','wheel-base']][10:60].plot.bar(figsize = (20,8), width = 0.8)
```

#### Histogram 
    * Continious Samples - Study the spread/ Distribution of Data.
    * Univariate Analysis.

```python
fig = plt.figure(figsize = (10,6))
# bins = intervals
plt.hist(auto_price['price'], color = 'lightgreen', bins = 15)
plt.title('Histogram Plot - Auto Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

#### Subplots

```python
# Subplots = (1, 2, 1) => 1st Row, 2nd Column, 1st Index
plt.figure(figsize = (18,6))
plt.subplot(1, 2, 1)
plt.hist(auto_price['engine-size'], color = 'deepskyblue', bins = 20)
plt.title('Histogram Plot - Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Number of Vehicle')

plt.subplot(1, 2, 2)
plt.hist(auto_price['price'], color = 'lightgreen', bins = 15)
plt.title('Histogram Plot - Auto Price')
plt.xlabel('Price')
plt.ylabel('Number of Vehicle')
plt.show()
```

#### Scatter Plot
    * Scatter Plot is plotted between numerical values. 
    * Shows relationship b/w x & y plots

```python
plt.figure(figsize = (10,7))
# marker = changing the scatter plots style o,x,X,s,d,>,<,*,.,^
sns.scatterplot(x = auto_price['city-mpg'], y = auto_price['price'], color = 'crimson', marker = 's', alpha = 0.7)
plt.show()
```

```python
plt.figure(figsize = (10,7))
sns.scatterplot(x = auto_price['city-mpg'], y = auto_price['price'], hue = auto_price['fuel-type'], color = 'crimson')
plt.show()
```

#### Boxplot

    * Distribution of Sample
    * Five Point Summary - min , max, Q1, median, Q3
    * Outliers

```python
plt.figure(figsize = (10,5))
sns.boxplot(x = 'price', data = auto_price)
```

```python
plt.figure(figsize = (12,8))
sns.boxplot(x = 'drive-wheels', y = 'price', data = auto_price, palette = 'Dark2')
```

#### Distribution Plot
    * Histogram + Density
    * Probability Density Function - Continious

```python
plt.figure(figsize = (10,6))
# hist = False : Disable Histogram
sns.distplot(auto_price['price'], color = 'crimson', hist = False)
plt.show()
```

#### Pie-Chart

```python
auto_price['make'].value_counts()
```

```python
plt.figure(figsize = (10,10))
# Shows distribution % in upto 2 decimal Places
plt.pie(auto_price['make'].value_counts(), autopct = '%0.2f%%', labels = auto_price['make'].value_counts().index)
plt.show()
```
